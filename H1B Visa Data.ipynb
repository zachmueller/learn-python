```python
import pandas as pd
```

```python
filename = 'data/H-1B_FY14_Q4.xlsx'
```

```python
with open(filename, 'rb') as f:
    df = pd.read_excel(f, sheetname='H-1B_FY2014')
df.head()
```

```python
df.columns
```

```python
keep_list = ['STATUS', 'LCA_CASE_SUBMIT', 'DECISION_DATE', 
             'LCA_CASE_EMPLOYMENT_START_DATE', 'LCA_CASE_EMPLOYMENT_END_DATE',
             'LCA_CASE_EMPLOYER_NAME', 'LCA_CASE_EMPLOYER_CITY', 'LCA_CASE_EMPLOYER_STATE',
             'LCA_CASE_EMPLOYER_POSTAL_CODE', 'LCA_CASE_SOC_NAME', 'LCA_CASE_JOB_TITLE',
             'LCA_CASE_WAGE_RATE_FROM', 'LCA_CASE_WAGE_RATE_TO', 'LCA_CASE_WAGE_RATE_UNIT',
             'FULL_TIME_POS', 'TOTAL_WORKERS','LCA_CASE_WORKLOC1_CITY', 'LCA_CASE_WORKLOC1_STATE'] 
```

```python
df = df[keep_list]
```

```python
df[['STATUS', 'LCA_CASE_SUBMIT']].groupby('STATUS').count().sort('LCA_CASE_SUBMIT')
```

```python
df['DECISION_TIME']=df['DECISION_DATE'] - df['LCA_CASE_SUBMIT']
df.head()
```

```python
df['DECISION_TIME_DAYS']=df['DECISION_TIME'].dt.days
df['DECISION_TIME_DAYS'].head()
```

```python
print(df['DECISION_TIME_DAYS'].mean())
print(df['DECISION_TIME_DAYS'].max())
print(df['DECISION_TIME_DAYS'].min())
```

```python
df[['STATUS', 'DECISION_TIME_DAYS']].groupby('STATUS').mean()
```

```python
df[['LCA_CASE_EMPLOYER_STATE', 'LCA_CASE_WAGE_RATE_FROM', 'LCA_CASE_WAGE_RATE_UNIT']].groupby(['LCA_CASE_WAGE_RATE_UNIT', 'LCA_CASE_EMPLOYER_STATE']).mean()
```

```python
df[['LCA_CASE_WORKLOC1_STATE', 'LCA_CASE_WAGE_RATE_FROM', 'LCA_CASE_WAGE_RATE_UNIT']].groupby(['LCA_CASE_WAGE_RATE_UNIT', 'LCA_CASE_WORKLOC1_STATE']).mean()
```

```python
df_subset[df_subset['LCA_CASE_EMPLOYER_NAME']=='IBM CORPORATION'].sort('LCA_CASE_WAGE_RATE_FROM')
```

```python
df[(df['STATUS']=='CERTIFIED')&(df['LCA_CASE_WAGE_RATE_UNIT']=='Year')][['LCA_CASE_WORKLOC1_STATE', 'LCA_CASE_WAGE_RATE_FROM', 'LCA_CASE_WAGE_RATE_UNIT']].groupby(['LCA_CASE_WAGE_RATE_UNIT', 'LCA_CASE_WORKLOC1_STATE']).mean().sort('LCA_CASE_WAGE_RATE_FROM')
```

```python
df_subset = df[(df['STATUS']=='CERTIFIED')& (df['LCA_CASE_WORKLOC1_STATE']=='WY') & (df['LCA_CASE_WAGE_RATE_UNIT']=='Year')]
df_subset[['LCA_CASE_JOB_TITLE', 'LCA_CASE_WAGE_RATE_FROM']].groupby('LCA_CASE_JOB_TITLE').mean().sort('LCA_CASE_WAGE_RATE_FROM')
```

```python
total_applications_by_state = df[['STATUS', 'LCA_CASE_WORKLOC1_STATE']].groupby('LCA_CASE_WORKLOC1_STATE').count()
total_applications_by_state = total_applications_by_state.reset_index()
total_applications_by_state.columns = ['STATE', 'TOTAL_COUNT']
total_applications_by_state
```

```python
accepted_applications_by_state = df[(df['STATUS'] == 'CERTIFIED') | (df['STATUS'] == 'CERTIFIED-WITHDRAWN')][['STATUS', 'LCA_CASE_WORKLOC1_STATE']].groupby('LCA_CASE_WORKLOC1_STATE').count()
accepted_applications_by_state = accepted_applications_by_state.reset_index()
accepted_applications_by_state.columns = ['STATE', 'ACCEPTED_COUNT']
accepted_applications_by_state
```

```python
df_merge=pd.merge(left=total_applications_by_state, right=accepted_applications_by_state, how='inner', on='STATE')
df_merge['ACCEPTANCE_RATE']=df_merge['ACCEPTED_COUNT']/df_merge['TOTAL_COUNT']
df_merge
```
